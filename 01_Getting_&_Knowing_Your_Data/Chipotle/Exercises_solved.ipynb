### Step 9. Which was the most-ordered item? 

```python
grouped_chipo = chipo.groupby("item_name")["quantity"].sum()
grouped_chipo.sort_values(ascending = False, inplace=True)
grouped_chipo
```

```python
grouped_chipo.index[0]
```

### Step 10. For the most-ordered item, how many items were ordered?

### Step 11. What was the most ordered item in the choice_description column?

```python
grouped_by_desc = chipo.groupby("choice_description")["choice_description"].count()
grouped_by_desc.sort_values(ascending=False, inplace=True)
grouped_by_desc
```

```python
grouped_by_desc.index[0]
```

### Step 13. Turn the item price into a float

#### Step 13.a. Check the item price type

```python
chipo["item_price"].dtype
```

#### Step 13.b. Create a lambda function and change the type of item price

```python
# note that there were currency symbol
chipo["item_price"] = chipo["item_price"].apply(lambda x: float(x[1:]) )
```

#### Step 13.c. Check the item price type

```python
chipo["item_price"].dtype
```

### Step 14. How much was the revenue for the period in the dataset?

```python
chipo["total_price"] = chipo["item_price"] * chipo["quantity"]
chipo["total_price"].sum()
```

### Step 15. How many orders were made in the period?

```python
# note that one order id can have several item
# group by order id
len(chipo.groupby("order_id"))
```

### Step 16. What is the average revenue amount per order?

```python
# Solution 1
group_by_id = chipo.groupby("order_id")["total_price"].sum()
sum(group_by_id) / len(chipo.groupby("order_id"))
```
